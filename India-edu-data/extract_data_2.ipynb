```python
import pandas as pd
```

```python
df = pd.read_csv("./raw-dataset/cleaned_LE.csv")
df
```

```python
male_columns = [col for col in df.columns if col.endswith("_Male")]
female_columns = [col for col in df.columns if col.endswith("_Female")]
both_columns = male_columns + female_columns

# Create DataFrames for each gender
male_df = df[["state"] + male_columns]
female_df = df[["state"] + female_columns]
both_df = df[["state"] + both_columns]

# Rename columns to have only years
male_df.columns = ["state"] + [col.split("_")[0] for col in male_columns]
female_df.columns = ["state"] + [col.split("_")[0] for col in female_columns]
both_df.columns = ["state"] + [col.split("_")[0] for col in both_columns]

```

```python
male_df.to_csv("le-india-male.csv", index=False)
female_df.to_csv("le-india-female.csv", index=False)
both_df.to_csv("le-india-both.csv", index=False)
```
